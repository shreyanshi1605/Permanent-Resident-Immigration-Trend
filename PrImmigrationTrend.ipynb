```python
import pandas as pd
idata=pd.read_excel("PR-ProvImmCat.xlsx")
```

```python
idata.info()
```

```python
idata=idata.dropna(how='any',axis=1)
```

```python
idata.head(15)
```

```python
idata=idata.replace('--','0')
```

```python
idata
```

```python
idata=idata.melt(id_vars=["Province/Territory and Immigration Category","Category","Sub-Category"],
                 var_name="Year",
                 value_name="No. of Permanent Resident")
```

```python
idata
```

```python
new = idata["Year"].str.split(" ",n=1,expand=True)
idata["Month"] = new[0]
idata["Year "] = new[1]
idata.drop(columns=["Year"],inplace=True)
```

```python
idata.index = np.arange(1,len(idata)+1)
```

```python
df = pd.DataFrame(idata)
```

```python
fig_dims = (15,10)
fig, ax = plt.subplots(figsize=fig_dims)
ax = sns.barplot(x="Year ", y="No. of Permanent Resident", data=idata, estimator=sum,palette="Blues_d").set_title("PR Immigrants by Year from 2015-2024")
plt.xlabel("Year", size=16)
plt.ylabel("No. of Permanent Resident",size=16)
sns.despine()
```
